```python
import pandas as pd
```

```python
mvps = pd.read_csv('mvp.csv')
```

```python
mvps
```

```python
mvps = mvps[["Player",'Year', 'Pts Won', 'Pts Max', 'Share']]
```

```python
mvps.head()
```

```python
players = pd.read_csv('players.csv')
```

```python
players
```

```python
players.columns
```

```python
del players['Rk']
```

```python
players['Player'] = players['Player'].str.replace('*', "", regex=False)
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Team'] == 'TOT']
        row['Team'] = df.iloc[-1,:]['Team']
        return row

players = players.groupby(['Player', 'Year']).apply(single_row)
```

```python
players
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```

```python
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python

combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined
```

```python
mvps_to_fill = ['Pts Won','Pts Max','Share']
```

```python
combined[mvps_to_fill] = combined[mvps_to_fill].fillna(0)
```

```python
combined.head()
```

```python
del combined['Awards']
```

```python
combined.head()
```

```python
teams = pd.read_csv('teams.csv')
```

```python
teams.head()
```

```python
teams = teams[~teams['W'].str.contains('Division')]
```

```python
teams['Team'] = teams['Team'].str.replace('*', '', regex = False)
```

```python
teams.head(30)
```

```python
del teams['Unnamed: 0']
```

```python
with open('nicknames.csv') as f:
    lines = f.readlines()
lines
```

```python
nicknames = {}

with open('nicknames.csv') as f:
    lines = f.readlines()

    for line in lines[1:]: #skipping the first row coz its the tittle
        Team, Abbreviation  = line.replace('\n','').split(',')
        nicknames[Abbreviation] = Team
```

```python
combined['Team'] = combined['Team'].map(nicknames)
```

```python

combined.head(5)
```

```python
stats = combined.merge(teams, how='outer',on=(['Team', 'Year']))
```

```python
stats.tail(100)
```

```python
stats.dtypes
```

```python

stats = stats.apply(pd.to_numeric, errors='ignore')
```

```python
stats.dtypes
```

```python
stats['GB'].unique()
```

```python
stats['GB'] = stats['GB'].str.replace('—', '0')
```

```python
stats['GB'] = pd.to_numeric(stats['GB'])
```

```python
stats = stats.fillna(0)
```

```python
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```

```python
highest_scoring_by_year = stats.groupby("Year").apply(lambda x: x.sort_values("PTS", ascending=False).head(1))
```
